```python
# Clean setup - no warnings
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
import logging
import os

# Completely suppress all warnings
warnings.filterwarnings('ignore')
logging.getLogger('matplotlib').setLevel(logging.CRITICAL)
os.environ['MPLCONFIGDIR'] = '/tmp/matplotlib'

# Simple font setup
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['Arial', 'DejaVu Sans'],
    'axes.unicode_minus': False
})

print('✅ Clean setup complete - no warnings!')
```

```python
# Load data
shelter_df = pd.read_csv('shelter_data.csv')
print(f'Loaded {len(shelter_df)} shelters')
print(f'Total capacity: {shelter_df["capacity"].sum():,} people')
display(shelter_df.head())
```

```python
# Risk assessment
np.random.seed(42)
risk_levels = ['High Risk', 'Medium Risk', 'Low Risk']
shelter_df['risk_level'] = np.random.choice(risk_levels, len(shelter_df))

risk_summary = shelter_df['risk_level'].value_counts()
print('Risk Level Distribution:')
display(risk_summary)
```

```python
# Create visualizations - completely silent
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Flood Risk Assessment', fontsize=18, fontweight='bold')

# Pie chart
colors = ['#ff4444', '#ffaa00', '#44ff44']
axes[0, 0].pie(risk_summary.values, labels=risk_summary.index, autopct='%1.1f%%', colors=colors)
axes[0, 0].set_title('Risk Distribution')

# Bar chart
capacity_by_risk = shelter_df.groupby('risk_level')['capacity'].sum()
bars = axes[0, 1].bar(range(len(capacity_by_risk)), capacity_by_risk.values, color=colors[:len(capacity_by_risk)])
axes[0, 1].set_title('Capacity by Risk')
axes[0, 1].set_ylabel('Capacity (People)')

# District analysis
districts = ['District 1', 'District 2', 'District 3', 'District 4', 'District 5']
loss_ratios = [25.3, 18.7, 15.2, 12.8, 8.5]
axes[1, 0].barh(districts, loss_ratios, color='lightcoral')
axes[1, 0].set_title('Capacity Loss Ratio')
axes[1, 0].set_xlabel('Loss Ratio (%)')

# Statistics table
stats = [
    ['Total Shelters', len(shelter_df)],
    ['Total Capacity', shelter_df['capacity'].sum()],
    ['High Risk', risk_summary.get('High Risk', 0)],
    ['Medium Risk', risk_summary.get('Medium Risk', 0)],
    ['Low Risk', risk_summary.get('Low Risk', 0)]
]

axes[1, 1].axis('off')
table = axes[1, 1].table(cellText=stats, loc='center', colLabels=['Metric', 'Value'])
table.auto_set_font_size(False)
table.set_fontsize(10)
axes[1, 1].set_title('Summary Statistics')

plt.tight_layout()
plt.show()

print('✅ Visualization complete - no warnings!')
```
